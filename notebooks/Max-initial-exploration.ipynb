## Feature Engineer *Credit Utilization Ratio* and Bin Ratios

```python
one_loan_credit_card_df = credit_card_balance_df.query("SK_ID_CURR == 371185")
```

```python
one_loan_credit_card_df["Credit Utilization Ratio"] = one_loan_credit_card_df["AMT_BALANCE"] / one_loan_credit_card_df["AMT_CREDIT_LIMIT_ACTUAL"]
```

```python
one_loan_credit_card_df.head()
```

```python

labels = ["0.00 - 0.25", "0.25 - 0.50", "0.50 - 0.75", "0.75 - 1.00"]

one_loan_credit_card_df["Credit Utilization Ratio"] = pd.cut(one_loan_credit_card_df["Credit Utilization Ratio"], bins=[0, .25, .5, .75, 1.], labels=labels)
```

## One Hot Encode Ratio with n/a values

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
ohe.fit(one_loan_credit_card_df[["Credit Utilization Ratio"]])
ratio_encoded = ohe.transform(one_loan_credit_card_df[["Credit Utilization Ratio"]])
ratio_encoded
```

```python
one_loan_credit_card_df[labels[0]], one_loan_credit_card_df[labels[1]], one_loan_credit_card_df[labels[2]], one_loan_credit_card_df[labels[3]], one_loan_credit_card_df["n/a"] = ratio_encoded.T  
```
